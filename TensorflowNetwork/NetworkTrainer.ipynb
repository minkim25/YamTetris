```python
def prettyPrintTetris(tetrisMove):
    row = ''
    acc = 1
    columns = []
    for box in tetrisMove[8:228]:
        if acc != 10:
            row+=str(box)
            acc += 1
        else: # acc == 10
            row+=str(box)
            columns.append(row)
            row = ''
            acc = 1
    [print(x) for x in columns]
```
